#1. Cleaning Overview
Based on data understanding stage, the problem with this dataset is duplicated data and column name

```python
#2. Load Dataset 
import pandas as pd 

# Data source
df_raw = pd.read_csv('../data/raw_Daily_Water_Intake.csv')
```

```python
#3. Clean Duplicated Data

# Validate the duplicated data
data_duplicated = df_raw.duplicated(keep=False)
df_raw[data_duplicated].sort_values(df_raw.columns.tolist())

# Drop duplicate data and use the last one, because this dataset didn't provide explicit timestamp
df_single = df_raw.drop_duplicates(keep='last')

# Check table shape
df_single.shape
```

```python
#4. Column Name Correction

# Checkking column names
df_single.head()

# Convert column names to lowercase, replace space with underscore and delete parentheses 
df_single.columns = (
    df_single.columns
    .str.strip()
    .str.replace(' ','_')
    .str.replace(r'[()]','',regex=True)
    .str.lower()
)

# Assign new dataframe to df_clean and print
df_clean = df_single.copy()
df_clean.head()
```

Every duplicated rows were already identified and deleted from the dataset. The last data recorded from duplicated data is used based on file ordering, as no timestamp column was provided in this dataset.

```python
#5. Export clean data
df_clean.to_csv('../data/clean_Daily_Water_Intake.csv', index=False)
```
