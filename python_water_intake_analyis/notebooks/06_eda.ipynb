#1. This data exploratory processs is to understand relationship, distribution and uncover insight related to daily water intake.

```python
# Load Clean Data and Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import shapiro, normaltest

df_clean = pd.read_csv('../data/clean_Daily_Water_Intake.csv')
df_clean.head()
```

1. What is the average of daily water consumption per age group?

```python
# Average daily water consumption per age group
avg_water_age = df_clean.groupby("age_group")["daily_water_intake_liters"].mean()
print("Rata-rata Konsumsi Air Harian Berdasarkan Kelompok Usia:")
print(avg_water_age)

# Visualize average daily water consumption per age group    
ax = avg_water_age.plot(kind='bar',title='Average Daily Water Intake by Age Group', ylabel='Liters', xlabel='Age Group')

# Add value labels on top of bars
for container in ax.containers:
    ax.bar_label(container)

```

2. Is daily water consumption normally distributed?

```python
# print("Mean of Daily Water Intake:")
# print(df_clean['daily_water_intake_liters'].mean())
# print("Standard Deviation of Daily Water Intake:")
print(df_clean['daily_water_intake_liters'].std())

plt.hist(df_clean['daily_water_intake_liters'], bins=30)
plt.show()




stats.probplot(df_clean["daily_water_intake_liters"].dropna(), dist="norm", plot=plt)
plt.show()


stat, p = shapiro(df_clean["daily_water_intake_liters"].dropna())
print("p-value:", p)

stat, p = normaltest(df_clean["daily_water_intake_liters"].dropna())
print("p-value:", p)
```
